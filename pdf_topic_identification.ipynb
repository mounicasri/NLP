```python
import PyPDF2
```

```python
article_list = []
with open("\\IMPROVING-SENTIMENT-ANALYSIS-WITH-PART-OF-SPEECH-05212278.pdf",'rb') as file:
    pdf = PyPDF2.PdfFileReader(file)
    page_count = pdf.getNumPages()
    for i in range(0,page_count):
        page_obj = pdf.getPage(i)
        article_list.append(page_obj.extractText())
    article = ' '.join(article_list)
article
```

```python
#Tokenize
tokens = word_tokenize(article)
tokens
```

```python
#remove special characters
art_alpha = [word for word in tokens if word.isalpha()]
art_alpha
```

```python
art_alpha_lower = [word.lower() for word in art_alpha]
art_alpha_lower
```

```python
art_alpha_lower_stop = [word for word in art_alpha_lower if word not in stopwords.words('english')]
art_alpha_lower_stop
```

```python
word_lemma = WordNetLemmatizer()
art_alpha_lower_stop_lemma = [word_lemma.lemmatize(word) for word in art_alpha_lower_stop]
art_alpha_lower_stop_lemma
```

```python
from collections import Counter
counts = Counter(art_alpha_lower_stop_lemma)
counts.most_common(10)
```
